```python
# Ячейка 1 — импорты и настройки
%matplotlib inline
import re
from pathlib import Path
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
from collections import Counter

INPUT_XLSX = Path("Бровки_now.xlsx")   # <-- поменяй путь если нужно
assert INPUT_XLSX.exists(), f"Файл не найден: {INPUT_XLSX}"
OUTPUT_DIR = Path("results_models")
OUTPUT_DIR.mkdir(exist_ok=True)

# Параметры (жёсткие, под твой формат)
HEADER_ROW_IDX = 1      # индекс строки с названиями столбцов (вторая строка Excel)
DATA_START_ROW = 2      # индекс строки, где начинаются данные (третья строка Excel)
COLS_PER_PROFILE = 5    # число колонок в каждом профиле
```

```python
# Ячейка 2 — утилиты очистки
def safe_str(x):
    return "" if pd.isna(x) else str(x).strip()

def clean_num_series(s):
    return pd.to_numeric(
        s.astype(str)
         .str.replace(r'\s+', '', regex=True)
         .str.replace('*', '', regex=False)
         .str.replace(',', '.', regex=False)
         .str.replace('−', '-', regex=False)
         .str.replace(r'[^0-9\.\-]', '', regex=True),
        errors='coerce'
    )
```

```python
# Ячейка 3 — функция парсинга одного листа (жёсткая разбивка по блокам по COLS_PER_PROFILE)
def parse_sheet_fixed(sheet_df, header_row_idx=HEADER_ROW_IDX, data_start_row=DATA_START_ROW, cols_per_profile=COLS_PER_PROFILE):
    ncols = sheet_df.shape[1]
    n_profiles = ncols // cols_per_profile
    profiles = {}
    combined_parts = []
    # попытаемся взять "заголовок профиля" из строки 0 (если там есть "ПРОФИЛЬ № 1" и т.д.)
    title_row = sheet_df.iloc[0] if sheet_df.shape[0] > 0 else None

    # стандартные имена заголовков внутри блока (как в Excel)
    expected_headers = ['Дата измерения', 'ГП', 'ИГП, м', 'РГП-ПН, м', 'ИПН, м']
    # соответствие кода к коротким именам
    rename_map = {
        'Дата измерения': 'Дата',
        'Дата': 'Дата',
        'ГП': 'ГП',
        'ИГП, м': 'ИГП_м',
        'ИГП': 'ИГП_м',
        'РГП-ПН, м': 'РГП_ПН_м',
        'РГП-ПН': 'РГП_ПН_м',
        'ИПН, м': 'ИПН_м',
        'ИПН': 'ИПН_м'
    }

    for i in range(n_profiles):
        start = i * cols_per_profile
        end = start + cols_per_profile
        block = sheet_df.iloc[:, start:end].copy()
        # Получаем имена колонок из header_row_idx (вторая строка)
        raw_colnames = block.iloc[header_row_idx].astype(str).fillna("").tolist()
        # Если raw_colnames пусты (например Excel с merged headers), используем expected_headers
        if all(safe_str(x) == "" for x in raw_colnames):
            raw_colnames = expected_headers[:block.shape[1]]
        # Если количество реальных колонок < cols_per_profile, подрубаем имена по позиции
        if len(raw_colnames) < block.shape[1]:
            # дополнить по ожидаемым
            for k in range(block.shape[1]):
                if k >= len(raw_colnames) or safe_str(raw_colnames[k]) == "":
                    raw_colnames = raw_colnames[:k] + [expected_headers[k]] + raw_colnames[k+1:]
        # нормализуем имена (замена в rename_map при наличии)
        norm_names = []
        for nm in raw_colnames:
            key = safe_str(nm)
            key_low = key.lower()
            mapped = None
            for candidate, short in rename_map.items():
                if candidate.lower() == key_low:
                    mapped = short; break
            if mapped is None:
                # попытка хитрой map: убрать пробелы и сравнить по содержимому
                key_s = re.sub(r'\s+','', key_low)
                if 'дата' in key_s:
                    mapped = 'Дата'
                elif 'игп' in key_s:
                    mapped = 'ИГП_м'
                elif 'ипн' in key_s or 'бров' in key_s:
                    mapped = 'ИПН_м'
                elif 'ргп' in key_s or 'пн' in key_s:
                    mapped = 'РГП_ПН_м'
                elif 'гп' == key_s:
                    mapped = 'ГП'
                else:
                    mapped = key  # оставим как есть
            norm_names.append(mapped)
        # сделаем уникальные имена, если вдруг повторяются
        from collections import Counter
        cnt = Counter()
        uniq_names = []
        for nm in norm_names:
            if cnt[nm] == 0:
                uniq_names.append(nm)
            else:
                uniq_names.append(f"{nm}__{cnt[nm]}")
            cnt[nm] += 1

        block.columns = uniq_names

        # Данные начинаются со строки data_start_row (индекс)
        data_block = block.iloc[data_start_row:].reset_index(drop=True).copy()

        # Привести типы
        # Дата
        date_col = None
        for c in data_block.columns:
            if 'дата' in c.lower():
                date_col = c; break
        if date_col is None:
            # попытка найти первую колонку, распознаваемую как дата
            for c in data_block.columns[:min(3, data_block.shape[1])]:
                if pd.to_datetime(data_block[c], errors='coerce', dayfirst=True).notna().sum() > 0:
                    date_col = c; break
        if date_col:
            data_block['Дата'] = pd.to_datetime(data_block[date_col], errors='coerce', dayfirst=True)
        else:
            data_block['Дата'] = pd.NaT

        # Числовые
        if 'ИГП_м' in data_block.columns:
            data_block['ИГП_м'] = clean_num_series(data_block['ИГП_м'])
        else:
            # возможно имя с запятыми/пробелами - попробуем найти по частичному совпадению
            for c in data_block.columns:
                if 'игп' in c.lower() or 'измер' in c.lower():
                    data_block['ИГП_м'] = clean_num_series(data_block[c]); break

        if 'РГП_ПН_м' in data_block.columns:
            data_block['РГП_ПН_м'] = clean_num_series(data_block['РГП_ПН_м'])
        else:
            for c in data_block.columns:
                if 'ргп' in c.lower() or ('пн' in c.lower() and 'ргп' not in c.lower()):
                    data_block['РГП_ПН_м'] = clean_num_series(data_block[c]); break

        if 'ИПН_м' in data_block.columns:
            data_block['ИПН_м'] = clean_num_series(data_block['ИПН_м'])
        else:
            for c in data_block.columns:
                if 'ипн' in c.lower() or 'бров' in c.lower():
                    data_block['ИПН_м'] = clean_num_series(data_block[c]); break

        # ГП и Пункт (строки)
        if 'ГП' in data_block.columns:
            data_block['ГП'] = data_block['ГП'].astype(str).str.strip().replace("nan","")
        else:
            # попытка найти колонку с нетипичной меткой
            for c in data_block.columns:
                if c.lower().strip().startswith('гп'):
                    data_block['ГП'] = data_block[c].astype(str).str.strip().replace("nan",""); break

        # удалить строки без даты
        if 'Дата' in data_block.columns:
            data_block = data_block.dropna(subset=['Дата']).reset_index(drop=True)
        else:
            data_block = data_block.iloc[0:0]

        # профильное имя: берём из title_row если там есть текст в start_col, иначе Профиль_i
        profile_title = None
        try:
            if title_row is not None:
                v = safe_str(title_row.iloc[start])
                if v != "":
                    profile_title = v
        except Exception:
            profile_title = None
        if not profile_title:
            profile_title = f"Профиль_{i+1}"

        # Добавляем в словарь
        # нормализуем колонки: оставим только нужные короткие имена (если есть)
        keep = []
        for c in ['Дата','ГП','ИГП_м','РГП_ПН_м','ИПН_м']:
            if c in data_block.columns:
                keep.append(c)
        # иногда столбец 'Пункт' отсутствует; оставляем что есть
        if 'Пункт' in data_block.columns:
            keep.insert(1,'Пункт')  # после 'Дата'
        # reindex preserve order if present
        df_profile = data_block[[c for c in ['Дата','Пункт','ГП','ИГП_м','РГП_ПН_м','ИПН_м'] if c in data_block.columns]].copy()
        # добавим колонку Год
        if 'Дата' in df_profile.columns:
            df_profile['Год'] = df_profile['Дата'].dt.year
        # save
        profiles[profile_title] = df_profile
        combined_part = df_profile.copy()
        combined_part['Профиль'] = profile_title
        combined_parts.append(combined_part)

    # объединённый DF для листа
    combined = pd.concat(combined_parts, ignore_index=True) if combined_parts else pd.DataFrame()
    return profiles, combined
```

```python
# Ячейка 4 — пройти по книге, собрать sites_dict и итоговые таблицы
xls = pd.ExcelFile(INPUT_XLSX)
sites_dict = {}
for sheet in xls.sheet_names:
    raw = pd.read_excel(INPUT_XLSX, sheet_name=sheet, header=None)
    profiles, combined = parse_sheet_fixed(raw)
    # добавим колонку Участок в combined
    if not combined.empty:
        combined['Участок'] = sheet
    sites_dict[sheet] = {'profiles': profiles, 'combined': combined}

# объединённый DF со всеми наблюдениями
df_observations = pd.concat([v['combined'] for v in sites_dict.values() if not v['combined'].empty], ignore_index=True)
df_observations.to_excel(OUTPUT_DIR/"df_observations_simple.xlsx", index=False)
print("Участков:", len(sites_dict), "Всего наблюдений:", len(df_observations))
```

```python
# Ячейка 5 — сформировать df_profiles с рассчитанным retreat_m_per_year (линейный тренд)
def compute_slope_retreat(years, values):
    years = np.array(years, dtype=float)
    values = np.array(values, dtype=float)
    mask = ~np.isnan(years) & ~np.isnan(values)
    if mask.sum() < 2:
        return np.nan
    slope = np.polyfit(years[mask], values[mask], 1)[0]
    return -float(slope)  # положительная величина = отступание м/год

rows = []
for site, info in sites_dict.items():
    for prof_name, dfp in info['profiles'].items():
        if dfp.empty:
            continue
        # целевая: предпочитаем ИПН_м, иначе РГП_ПН_м
        if 'ИПН_м' in dfp.columns and dfp['ИПН_м'].notna().sum() >= 2:
            ycol = 'ИПН_м'
        elif 'РГП_ПН_м' in dfp.columns and dfp['РГП_ПН_м'].notna().sum() >= 2:
            ycol = 'РГП_ПН_м'
        else:
            ycol = None
        retreat = compute_slope_retreat(dfp['Год'], dfp[ycol]) if ycol else np.nan
        rows.append({
            'Участок': site,
            'Профиль': prof_name,
            'n_obs': int(len(dfp)),
            'year_first': int(dfp['Год'].min()),
            'year_last': int(dfp['Год'].max()),
            'ycol_used': ycol,
            'retreat_m_per_year': float(retreat) if not np.isnan(retreat) else np.nan
        })

df_profiles = pd.DataFrame(rows)
df_profiles.to_excel(OUTPUT_DIR/"df_profiles_simple_summary.xlsx", index=False)
print("Профилей в summary:", len(df_profiles))
df_profiles.head(12)
```
