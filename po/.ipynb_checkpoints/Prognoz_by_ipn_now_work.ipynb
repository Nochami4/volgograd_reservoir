```python
%matplotlib inline
import os
import re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Папка, где лежат отдельные xlsx-файлы по профилям (поправь если нужно)
PROFILES_DIR = Path("results_models/profiles")
assert PROFILES_DIR.exists(), f"Папка с профилями не найдена: {PROFILES_DIR}"

OUT_DIR = Path("results_models/profile_analysis")
(OUT_DIR / "plots" / "profiles").mkdir(parents=True, exist_ok=True)
(OUT_DIR / "plots" / "sites").mkdir(parents=True, exist_ok=True)
(OUT_DIR / "plots" / "corr").mkdir(parents=True, exist_ok=True)
(OUT_DIR / "corr_matrices").mkdir(parents=True, exist_ok=True)

print("Reading profile files from:", PROFILES_DIR)
print("Outputs will be saved to:", OUT_DIR)
# ---------- Ячейка 1: новые импорты и утилиты ----------
# Вставьте или замените этой ячейкой ваши глобальные импорты (в начале ноутбука).

import numpy as np
import pandas as pd

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# xgboost
# Если xgboost не установлен: pip install xgboost
#from xgboost import XGBRegressor

# tensorflow / keras для простого MLP
# Если tensorflow не установлен: pip install tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Утилиты
def compute_linear_slope(y_history, times=None):
    """
    y_history: np.array shape (n_samples, hist_len)
    times: array shape (hist_len,) или None -> np.arange(hist_len)
    Вернёт массив slope (n_samples,)
    """
    if y_history is None:
        raise ValueError("y_history is None")
    y_history = np.asarray(y_history)
    if times is None:
        times = np.arange(y_history.shape[1])
    def slope_one(yy):
        # если все значения Nan -> slope 0
        if np.all(np.isnan(yy)):
            return 0.0
        # линейная аппроксимация (полином степени 1)
        a, b = np.polyfit(times, yy, 1)
        return a
    slopes = np.apply_along_axis(slope_one, 1, y_history)
    return slopes

def safe_stack_horiz(arrs):
    """Горизонтально объединить массивы, которые могут быть 1D/2D"""
    arrs2 = []
    for a in arrs:
        a = np.asarray(a)
        if a.ndim == 1:
            a = a.reshape(-1,1)
        arrs2.append(a)
    return np.hstack(arrs2)
```

```python
def safe_str(x):
    return "" if pd.isna(x) else str(x).strip()

def clean_num_series(s):
    return pd.to_numeric(
        s.astype(str)
         .str.replace(r'\s+', '', regex=True)
         .str.replace('*', '', regex=False)
         .str.replace(',', '.', regex=False)
         .str.replace('−','-', regex=False)
         .str.replace(r'[^0-9\.\-]', '', regex=True),
        errors='coerce'
    )

def infer_site_profile_from_filename(p: Path):
    name = p.stem
    parts = name.split("__")
    if len(parts) >= 2:
        site = parts[0]
        profile = "__".join(parts[1:])
    else:
        m = re.match(r"(.+?)__?ПРОФИЛЬ_?(.+)", name, flags=re.I)
        if m:
            site = m.group(1)
            profile = m.group(2)
        else:
            if "профиль" in name.lower():
                site = name.split("профиль")[0].strip("_- ")
                profile = name
            else:
                site = name
                profile = name
    return safe_str(site), safe_str(profile)

# Соберём все xlsx в папке (и подпапках)
files = list(PROFILES_DIR.rglob("*.xls*"))
print("Found files:", len(files))

rows = []
file_table = []
for f in files:
    try:
        df = pd.read_excel(f, engine=None)
    except Exception as e:
        print("Failed to read", f, ":", e)
        continue

    site, profile = infer_site_profile_from_filename(f)
    col_map = {c: safe_str(c).lower() for c in df.columns}
    date_col = None
    ipn_col = None
    igp_col = None
    rgp_col = None
    gp_col = None
    punkt_col = None
    for c in df.columns:
        low = col_map[c]
        if 'дат' in low:
            date_col = c
        if 'ипн' in low or 'бров' in low:
            ipn_col = c
        if 'игп' in low or 'измер' in low:
            igp_col = c
        if 'ргп' in low or 'пн' in low:
            rgp_col = c
        if low.strip() == 'гп' or low.strip() == 'gp':
            gp_col = c
        if 'пункт' in low:
            punkt_col = c

    if date_col is None and df.shape[1] >= 1:
        date_col = df.columns[0]

    for idx, row in df.iterrows():
        try:
            if date_col is not None:
                date_val = pd.to_datetime(row[date_col], dayfirst=True, errors='coerce')
            else:
                date_val = pd.NaT
        except Exception:
            date_val = pd.NaT

        ipn_v = clean_num_series(pd.Series([row[ipn_col]]))[0] if ipn_col in df.columns else np.nan
        igp_v = clean_num_series(pd.Series([row[igp_col]]))[0] if igp_col in df.columns else np.nan
        rgp_v = clean_num_series(pd.Series([row[rgp_col]]))[0] if rgp_col in df.columns else np.nan
        gp_v = None
        if gp_col in df.columns:
            try:
                gp_v = safe_str(row[gp_col])
            except:
                gp_v = None
        punkt_v = None
        if punkt_col in df.columns:
            punkt_v = safe_str(row[punkt_col])

        if (pd.isna(date_val) and pd.isna(ipn_v) and pd.isna(igp_v) and pd.isna(rgp_v)):
            continue

        rows.append({
            'file': str(f),
            'site': site,
            'profile': profile,
            'date_raw': row[date_col] if date_col in df.columns else None,
            'Дата': date_val,
            'Год': int(date_val.year) if not pd.isna(date_val) else np.nan,
            'Пункт': punkt_v,
            'ГП': gp_v,
            'ИПН_м': ipn_v,
            'ИГП_м': igp_v,
            'РГП_ПН_м': rgp_v
        })
    file_table.append({'file': str(f), 'site': site, 'profile': profile, 'rows': df.shape[0]})

df_all = pd.DataFrame(rows)
print("Combined rows:", len(df_all))
pd.DataFrame(file_table).to_excel(OUT_DIR / "files_table.xlsx", index=False)
df_all.to_excel(OUT_DIR / "observations_from_files.xlsx", index=False)
df_all.head()
```

```python
# Ячейка 3 — нелинейный прогноз (полином 2-й степени) и улучшенная графика
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

def plot_profile_ipn_forecast(dfp, out_png, ycol='ИПН_м', title=None, degree=2):
    d = dfp.dropna(subset=['Год', ycol]).sort_values('Год')
    # Для полинома степени degree нужно минимум degree+1 точка, но для надёжности возьмём минимум 3
    if d.shape[0] < max(3, degree+1):
        return False
    
    X = d['Год'].values.reshape(-1, 1)
    y = d[ycol].values
    
    # Создаём модель полиномиальной регрессии
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    # Прогноз на 20 лет вперёд
    last_year = int(X.max())
    future_years = np.arange(last_year + 1, last_year + 21).reshape(-1, 1)
    X_all = np.vstack([X, future_years])
    y_all_pred = model.predict(X_all)
    
    # Построение графика
    plt.figure(figsize=(9, 5))
    
    # Фактические данные
    plt.scatter(X, y, color='steelblue', s=50, label='Фактические данные', zorder=3)
    plt.plot(X, y_pred, color='darkorange', linewidth=2, label=f'Полином {degree}-й степени (R² = {r2:.3f})')
    
    # Прогноз (пунктир)
    plt.plot(future_years, y_all_pred[-len(future_years):], 
             color='crimson', linestyle='--', linewidth=2, label='Прогноз на 20 лет')
    
    # Оформление
    plt.xlabel('Год', fontsize=12)
    plt.ylabel('ИПН, м', fontsize=12)
    plt.grid(True, linestyle=':', alpha=0.7)
    if title:
        plt.title(title, fontsize=14, fontweight='bold')
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.savefig(out_png, dpi=150, bbox_inches='tight')
    plt.close()
    return True
```

```python
def safe_filename(s: str) -> str:
    return re.sub(r'[^\w\d\-]', '_', str(s))[:120]

n_saved = 0
for (site, profile), grp in df_all.groupby(['site','profile']):
    out_name = OUT_DIR / "plots" / "profiles" / f"{safe_filename(site)}__{safe_filename(profile)}.png"
    title = f"{site} / {profile}"
    ok = plot_profile_ipn_forecast(grp, out_name, ycol='ИПН_м', title=title, degree=2)
    if ok:
        n_saved += 1
print("Profile forecast plots saved:", n_saved, "files ->", (OUT_DIR / "plots" / "profiles").resolve())
```

```python
# Ячейка 8 — графики временных рядов для сильно коррелированных пар профилей
min_corr = 0.7
n_pairs = 0

def shorten_profile_name(name, max_len=15):
    """Укорачивает название профиля: убирает 'ПРОФИЛЬ___', обрезает до max_len символов."""
    name = re.sub(r'ПРОФИЛЬ[_\s]*', '', name, flags=re.I)
    name = re.sub(r'профиль[_\s]*', '', name, flags=re.I)
    if len(name) > max_len:
        name = name[:max_len] + '...'
    return name

for site, g in df_all.groupby('site'):
    # Строим сводную таблицу: годы × профили, значения ИПН
    pivot = g.pivot_table(index='Год', columns='profile', values='ИПН_м', aggfunc='mean')
    pivot = pivot.dropna(axis=1, thresh=3)  # удаляем профили с <3 наблюдениями
    if pivot.shape[1] <= 1:
        continue
    
    # Корреляционная матрица для отбора пар
    corr = pivot.corr()
    cols = corr.columns.tolist()
    
    for i in range(len(cols)):
        for j in range(i+1, len(cols)):
            val = corr.iloc[i, j]
            if pd.isna(val) or abs(val) < min_corr:
                continue
            
            p1, p2 = cols[i], cols[j]
            # Выбираем данные по двум профилям, убираем пропуски
            df_pair = pivot[[p1, p2]].dropna()
            if df_pair.shape[0] < 3:
                continue
            
            # График временных рядов
            plt.figure(figsize=(10, 8))
            
            # Профиль 1 — синий
            plt.plot(df_pair.index, df_pair[p1], 'o-', color='steelblue', linewidth=2, markersize=6,
                     label=f'{shorten_profile_name(p1)} (синий)')
            # Профиль 2 — коричневый
            plt.plot(df_pair.index, df_pair[p2], 's-', color='saddlebrown', linewidth=2, markersize=6,
                     label=f'{shorten_profile_name(p2)} (коричневый)')
            
            plt.xlabel('Год', fontsize=12)
            plt.ylabel('ИПН, м', fontsize=12)
            plt.title(f'{site}: Сравнение профилей {shorten_profile_name(p1)} и {shorten_profile_name(p2)} (r={val:.2f})',
                      fontsize=14, fontweight='bold')
            plt.grid(True, linestyle=':', alpha=0.7)
            plt.legend(fontsize=11)
            plt.tight_layout()
            
            fn = OUT_DIR / "plots" / "corr" / f"{safe_filename(site)}__timeseries_{safe_filename(p1)}__{safe_filename(p2)}.png"
            plt.savefig(fn, dpi=150, bbox_inches='tight')
            plt.close()
            n_pairs += 1

print("Saved time series plots for strongly correlated profile pairs:", n_pairs)
```

```python
# Ячейка 8.5: интерполяция временных рядов для каждого профиля (без NaN)
from scipy.interpolate import interp1d

def interpolate_profile_safe(df_profile):
    """
    Возвращает DataFrame с колонками ['Год', 'ИПН_м'], где годы идут подряд
    с минимального по максимальный год, пропуски заполнены линейной интерполяцией.
    Если данных меньше 2 точек, заполняет все годы единственным значением (если есть).
    """
    if len(df_profile) == 0:
        return pd.DataFrame(columns=['Год', 'ИПН_м'])
    
    years = df_profile['Год'].values
    values = df_profile['ИПН_м'].values
    full_years = np.arange(years.min(), years.max() + 1)
    
    if len(df_profile) == 1:
        # Одна точка – копируем значение на все годы
        full_values = np.full(len(full_years), values[0])
    else:
        # Интерполяция с экстраполяцией
        try:
            f = interp1d(years, values, kind='linear', fill_value='extrapolate')
            full_values = f(full_years)
        except Exception:
            # Если интерполяция не удалась (например, из-за дубликатов), используем последнее значение
            full_values = np.full(len(full_years), values[-1])
        # Принудительно заменяем возможные NaN (если экстраполяция не удалась)
        full_values = np.nan_to_num(full_values, nan=values[-1])
    
    df_full = pd.DataFrame({'Год': full_years, 'ИПН_м': full_values})
    return df_full

# Применяем ко всем профилям
interpolated_dfs = []
for (site, profile), grp in df_all.groupby(['site', 'profile']):
    df_int = interpolate_profile_safe(grp.sort_values('Год'))
    if len(df_int) == 0:
        continue
    df_int['site'] = site
    df_int['profile'] = profile
    interpolated_dfs.append(df_int)

df_interp = pd.concat(interpolated_dfs, ignore_index=True)

# Финальная проверка и заполнение оставшихся NaN (исправлено: используем ffill/bfill без method)
def fill_na_group(grp):
    # grp - Series, применяем цепочно
    return grp.ffill().bfill().fillna(0)

df_interp['ИПН_м'] = df_interp.groupby(['site', 'profile'])['ИПН_м'].transform(fill_na_group)

print(f"После интерполяции: {len(df_interp)} строк")
print(f"NaN в ИПН_м: {df_interp['ИПН_м'].isna().sum()}")
```

```python
# Ячейка 9: подготовка последовательностей (окна значений ИПН + год) для каждого профиля
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from pathlib import Path

# Параметры
LOOKBACK = 15          # сколько прошлых лет используем
N_FUTURE = 20         # прогноз на 20 лет вперёд
MIN_POINTS = LOOKBACK + 3  # минимум точек для профиля

# Словари для хранения данных по профилям
models = {}           # модель нейросети
scalers_ipn = {}      # StandardScaler для ИПН (нормализация значений)
scalers_year = {}     # StandardScaler для года (нормализация года)
profile_info = {}     # метаданные: последние значения, последний год и т.д.

profiles = df_interp.groupby(['site', 'profile']).size().reset_index(name='count')
print(f"Всего профилей после интерполяции: {len(profiles)}")
```

```python
# Ячейка 10 (исправленная): функция для создания обучающих последовательностей с защитой от нулевого std
def create_sequences_with_year(df_profile, lookback, future_years_max):
    """
    df_profile: DataFrame с колонками ['Год', 'ИПН_м'] (отсортирован по году)
    future_years_max: максимальный год прогноза (нужен для нормализации года)
    Возвращает:
        X: массив признаков (samples, lookback + 1) – последние lookback значений ИПН + нормализованный год
        y: целевое значение ИПН (следующий год)
        ipn_scaler: StandardScaler для ИПН (обучен на всех значениях профиля)
        year_scaler: StandardScaler для года (обучен на всём диапазоне лет от min до max_year)
        last_values: последние lookback значений ИПН (для старта прогноза)
        last_year: последний год наблюдений
    """
    years = df_profile['Год'].values
    values = df_profile['ИПН_м'].values

    # Нормализация ИПН с защитой от нулевого стандартного отклонения
    ipn_scaler = StandardScaler()
    ipn_scaler.fit(values.reshape(-1, 1))
    if ipn_scaler.scale_[0] < 1e-10:
        # Если все значения одинаковы, масштабирование не нужно: устанавливаем scale=1, mean=0
        ipn_scaler.scale_[0] = 1.0
        ipn_scaler.mean_[0] = 0.0
        ipn_scaler.var_[0] = 1.0

    # Нормализация года (от первого года до последнего года прогноза)
    first_year = years[0]
    last_obs_year = years[-1]
    max_year = max(last_obs_year, future_years_max)
    all_years = np.arange(first_year, max_year + 1).reshape(-1, 1)
    year_scaler = StandardScaler()
    year_scaler.fit(all_years)
    if year_scaler.scale_[0] < 1e-10:
        year_scaler.scale_[0] = 1.0
        year_scaler.mean_[0] = 0.0
        year_scaler.var_[0] = 1.0

    # Масштабируем значения
    values_scaled = ipn_scaler.transform(values.reshape(-1, 1)).flatten()
    years_scaled = year_scaler.transform(years.reshape(-1, 1)).flatten()

    X, y = [], []
    for i in range(len(values) - lookback):
        feat = np.concatenate([values_scaled[i:i+lookback], [years_scaled[i+lookback]]])
        X.append(feat)
        y.append(values_scaled[i+lookback])

    last_values_scaled = values_scaled[-lookback:]
    last_year_scaled = years_scaled[-1]

    return (np.array(X), np.array(y),
            ipn_scaler, year_scaler,
            last_values_scaled, last_year_scaled, years[-1])
```

```python
# Ячейка 11: простая модель (вход: lookback значений + год)
def build_model(input_dim):
    model = Sequential([
        Dense(64, activation='relu', input_shape=(input_dim,)),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dropout(0.2),
        Dense(1)  # линейная активация для отрицательных значений
    ])
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
# Ячейка 12: цикл по профилям, обучение на последовательностях (ИПН + год)
for (site, profile), grp in df_interp.groupby(['site', 'profile']):
    df_prof = grp.sort_values('Год').reset_index(drop=True)

    if len(df_prof) < MIN_POINTS:
        print(f"Пропущен {site} / {profile}: точек {len(df_prof)} < {MIN_POINTS}")
        continue

    # Определяем максимальный год прогноза для нормализации
    last_obs_year = df_prof['Год'].iloc[-1]
    future_max = last_obs_year + N_FUTURE

    # Создаём последовательности
    result = create_sequences_with_year(df_prof, LOOKBACK, future_max)
    if result[0] is None or len(result[0]) < 3:
        print(f"Пропущен {site} / {profile}: недостаточно окон")
        continue

    X, y, ipn_scaler, year_scaler, last_vals_scaled, last_year_scaled, last_year = result

    # Проверяем last_vals_scaled на NaN
    if np.any(np.isnan(last_vals_scaled)):
        print(f"Пропущен {site} / {profile}: last_vals_scaled содержит NaN")
        continue

    # Разделение на train/val (80/20)
    split = int(0.8 * len(X))
    if split < 1 or split >= len(X):
        X_train, y_train = X, y
        X_val, y_val = X[:0], y[:0]
    else:
        X_train, X_val = X[:split], X[split:]
        y_train, y_val = y[:split], y[split:]

    # Модель
    model = build_model(LOOKBACK + 1)  # +1 за счёт года
    early_stop = EarlyStopping(monitor='val_loss' if len(X_val) > 0 else 'loss',
                               patience=15, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val) if len(X_val) > 0 else None,
        epochs=300,
        batch_size=min(8, len(X_train)),
        callbacks=[early_stop] if len(X_val) > 0 else None,
        verbose=0
    )

    # Остатки для доверительного интервала (на обучающей выборке)
    y_train_pred = model.predict(X_train, verbose=0).flatten()
    resid_std = np.std(y_train - y_train_pred)

    # Сохраняем всё
    key = (site, profile)
    models[key] = model
    scalers_ipn[key] = ipn_scaler
    scalers_year[key] = year_scaler
    profile_info[key] = {
        'last_vals_scaled': last_vals_scaled,
        'last_year_scaled': last_year_scaled,
        'last_year': last_year,
        'resid_std': resid_std
    }
    print(f"Обучили {site} / {profile}: окон {len(X)}")
```

```python
# Ячейка 13: прогноз для всех профилей с fallback на линейный тренд при ошибках нейросети
OUT_DIR = Path("results_models/profile_analysis")
SITE_PLOT_DIR = OUT_DIR / "plots" / "site_profiles_forecast_final"
SITE_PLOT_DIR.mkdir(parents=True, exist_ok=True)

def iterative_forecast_safe(model, ipn_scaler, year_scaler, last_vals_scaled, last_year_scaled, last_year, n_steps):
    """
    Пытается получить прогноз нейросети. Если на любом шаге возникает NaN,
    возвращает None.
    """
    preds_scaled = []
    cur_vals = last_vals_scaled.copy()
    cur_year_scaled = last_year_scaled
    cur_year = last_year

    for step in range(n_steps):
        if np.any(np.isnan(cur_vals)):
            return None

        next_year = cur_year + 1
        next_year_scaled = year_scaler.transform([[next_year]])[0, 0]
        X_input = np.concatenate([cur_vals, [next_year_scaled]]).reshape(1, -1)

        try:
            pred_scaled = model.predict(X_input, verbose=0).flatten()[0]
        except Exception:
            return None

        if np.isnan(pred_scaled):
            return None

        pred_scaled = np.clip(pred_scaled, -10, 10)  # защита от взрывов
        preds_scaled.append(pred_scaled)

        cur_vals = np.roll(cur_vals, -1)
        cur_vals[-1] = pred_scaled
        cur_year = next_year
        cur_year_scaled = next_year_scaled

    preds_scaled_arr = np.array(preds_scaled).reshape(-1, 1)
    try:
        preds_abs = ipn_scaler.inverse_transform(preds_scaled_arr).flatten()
    except Exception:
        return None

    if np.any(np.isnan(preds_abs)):
        return None

    return preds_abs

def linear_trend_forecast(df_profile, n_steps):
    """
    Строит линейный тренд по всем данным профиля и экстраполирует на n_steps вперёд.
    Возвращает года и предсказанные значения.
    """
    years = df_profile['Год'].values.reshape(-1, 1)
    values = df_profile['ИПН_м'].values
    lr = LinearRegression()
    lr.fit(years, values)
    last_year = years[-1][0]
    future_years = np.arange(last_year + 1, last_year + n_steps + 1).reshape(-1, 1)
    pred = lr.predict(future_years)
    return future_years.flatten(), pred

all_forecasts = []

for site, profiles_group in df_interp.groupby('site'):
    plt.figure(figsize=(14, 8))
    site_keys = [k for k in models.keys() if k[0] == site]
    if not site_keys:
        plt.close()
        continue

    colors = plt.cm.tab20(np.linspace(0, 1, len(site_keys)))

    for idx, (s, prof) in enumerate(site_keys):
        key = (s, prof)
        if key not in models or key not in scalers_ipn or key not in scalers_year or key not in profile_info:
            print(f"⚠️ Пропущен {site} / {prof}: отсутствуют данные")
            continue

        model = models[key]
        ipn_scaler = scalers_ipn[key]
        year_scaler = scalers_year[key]
        info = profile_info[key]

        required = ['last_vals_scaled', 'last_year_scaled', 'last_year', 'resid_std']
        if not all(f in info for f in required):
            print(f"⚠️ Пропущен {site} / {prof}: неполная информация {list(info.keys())}")
            continue

        if np.any(np.isnan(info['last_vals_scaled'])):
            print(f"⚠️ Пропущен {site} / {prof}: last_vals_scaled содержит NaN")
            # Рисуем только факт
            df_prof = df_interp[(df_interp['site'] == s) & (df_interp['profile'] == prof)].sort_values('Год')
            if len(df_prof) > 0:
                plt.plot(df_prof['Год'], df_prof['ИПН_м'], 'o-', color=colors[idx], linewidth=2,
                         label=f"{prof[:20]}..." if len(prof) > 20 else prof)
            continue

        df_prof = df_interp[(df_interp['site'] == s) & (df_interp['profile'] == prof)].sort_values('Год')
        if len(df_prof) == 0:
            print(f"⚠️ Пропущен {site} / {prof}: нет данных в df_interp")
            continue

        years_hist = df_prof['Год'].values
        values_hist = df_prof['ИПН_м'].values

        # Пробуем нейросетевой прогноз
        print(f"🔄 Прогноз (нейросеть) для {site} / {prof}...")
        pred_abs = iterative_forecast_safe(
            model, ipn_scaler, year_scaler,
            info['last_vals_scaled'],
            info['last_year_scaled'],
            info['last_year'],
            N_FUTURE
        )

        if pred_abs is None:
            # Fallback на линейный тренд
            print(f"⚠️ Нейросеть не удалась, используем линейный тренд для {site} / {prof}")
            try:
                future_years, pred_abs = linear_trend_forecast(df_prof, N_FUTURE)
                pred_years = future_years
                std_abs = 0  # без доверительного интервала
            except Exception as e:
                print(f"❌ Ошибка линейного тренда для {site} / {prof}: {e}")
                # Рисуем только факт
                plt.plot(years_hist, values_hist, 'o-', color=colors[idx], linewidth=2,
                         label=f"{prof[:20]}..." if len(prof) > 20 else prof)
                continue
        else:
            pred_years = np.arange(info['last_year'] + 1, info['last_year'] + N_FUTURE + 1)
            scale_ipn = ipn_scaler.scale_[0]
            std_abs = info['resid_std'] * scale_ipn

        # Сохраняем прогноз
        for yr, val in zip(pred_years, pred_abs):
            all_forecasts.append({'site': site, 'profile': prof, 'year': yr, 'pred_IPN': val})

        # Рисуем факт
        plt.plot(years_hist, values_hist, 'o-', color=colors[idx], linewidth=2,
                 label=f"{prof[:20]}..." if len(prof) > 20 else prof)

        # Рисуем прогноз
        plt.plot(pred_years, pred_abs, '--', color=colors[idx], linewidth=2)

        # Доверительный интервал (только для нейросети)
        if std_abs > 0:
            upper = pred_abs + std_abs * np.sqrt(np.arange(1, N_FUTURE + 1))
            lower = pred_abs - std_abs * np.sqrt(np.arange(1, N_FUTURE + 1))
            plt.fill_between(pred_years, lower, upper, color=colors[idx], alpha=0.15)

    # Легенда и оформление
    if plt.gca().lines:
        plt.legend(loc='best', fontsize=8)
    plt.xlabel('Год', fontsize=12)
    plt.ylabel('ИПН, м', fontsize=12)
    plt.title(f'Участок: {site} — все профили (прогноз с учётом года, fallback на тренд)', fontsize=14, fontweight='bold')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.5, alpha=0.5)
    plt.tight_layout()

    out_png = SITE_PLOT_DIR / f"{site.replace(' ', '_')}_final_forecast.png"
    plt.savefig(out_png, dpi=150)
    plt.close()
    print(f"✅ График участка сохранён: {out_png}")

df_forecasts = pd.DataFrame(all_forecasts)
df_forecasts.to_excel(OUT_DIR / "profiles_forecasts_final.xlsx", index=False)
print(f"\n📊 Прогнозы по профилям сохранены. Всего записей: {len(df_forecasts)}")
```

```python
# ---------- Ячейка A: агрегирование прогнозов по участкам и подготовка координат для наложения ----------
# Вставьте эту ячейку сразу после Ячейки 13, где формируется df_forecasts.

import numpy as np
import pandas as pd
from pathlib import Path
import json
from PIL import Image

MAP_PATH = "Карта.jpg"
OUT_DIR = Path("results_models/profile_analysis/map_overlay")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Проверки на существование переменных, которые были сформированы в предыдущих ячейках
if 'df_forecasts' not in globals():
    raise RuntimeError("В ноутбуке не найдена таблица df_forecasts — выполните ячейки до Ячейки 13, чтобы её получить.")

# df_forecasts: ожидаем колонки ['site','profile','year','pred_IPN']
print("Уникальные участки (site) в df_forecasts:", df_forecasts['site'].unique())

# Берём последнее наблюдение по каждому профилю из df_interp (df_interp есть в ноутбуке)
if 'df_interp' not in globals():
    raise RuntimeError("Не найден df_interp — в ноутбуке должна быть подготовленная таблица интерполированных наблюдений (df_interp).")

# Находим последнее (последний год) наблюдение для каждого (site, profile)
last_obs = (df_interp.sort_values(['site','profile','Год'])
                     .groupby(['site','profile'])['ИПН_м']
                     .last()
                     .reset_index()
                     .rename(columns={'ИПН_м': 'last_obs_IPN'}))

# Определим для каждого прогноза шаг ближайшего предсказанного года (первый прогнозный год для профиля)
first_forecast_year = (df_forecasts.groupby(['site','profile'])['year'].min().reset_index().rename(columns={'year':'first_forecast_year'}))
# Сопоставим прогнозы первого года с последним наблюдением
first_preds = df_forecasts.merge(first_forecast_year, on=['site','profile'], how='left')
first_preds = first_preds[first_preds['year'] == first_preds['first_forecast_year']]

# Соединяем с last_obs
first_preds = first_preds.merge(last_obs, on=['site','profile'], how='left')

# Вычисляем изменение (delta) относительно последнего наблюдения (первый прогнозный год)
first_preds['delta_IPN'] = first_preds['pred_IPN'] - first_preds['last_obs_IPN']

# Агрегируем по участку (site) — среднее изменение, стандартное отклонение и число профилей
site_metrics = (first_preds.groupby('site')
                .agg(mean_delta=('delta_IPN','mean'),
                     std_delta=('delta_IPN','std'),
                     n_profiles=('delta_IPN','count'))
                .reset_index())

# Заполним NaN std = 0
site_metrics['std_delta'] = site_metrics['std_delta'].fillna(0.0)

print("Агрегированные метрики по участкам (первые 20):")
display(site_metrics.head(20))

# --- Подготовим координаты для наложения на растровую карту ---
# Если у вас есть точные пиксельные координаты для участков — заполните словарь site_coords_manual = {'Название участка': (x_pixel, y_pixel), ...}
site_coords_manual = {
    # Пример:
    # "г. Волгоград": (50, 180),
    # "г. Волжский": (70, 220),
    # Заполните вручную ваши соответствия, если есть
}

# Загрузим карту, чтобы знать размеры
img = Image.open(MAP_PATH).convert('RGBA')
W, H = img.size
print(f"Карта загружена: {MAP_PATH} размер {W}x{H} пикселей")

sites = site_metrics['site'].tolist()

# Если пользователь не заполнил координаты вручную — создадим приближённое распределение вдоль вертикали
site_coords = {}
missing_sites = [s for s in sites if s not in site_coords_manual]
if len(missing_sites) > 0:
    # разместим автоматически: X около центра карты (W*0.55), Y - равномерно распределим по диапазону [H*0.1, H*0.9]
    x_auto = int(W * 0.55)
    ys = np.linspace(int(H*0.12), int(H*0.88), len(missing_sites)).astype(int)
    for s, y in zip(missing_sites, ys):
        site_coords[s] = (x_auto, int(y))
# дополним вручную заданные
for s, coord in site_coords_manual.items():
    site_coords[s] = coord

# Сохраним mapping и метрики для последующей ручной правки / использования
(site_metrics.to_csv(OUT_DIR / "site_metrics.csv", index=False))
with open(OUT_DIR / "site_coords_suggested.json", 'w', encoding='utf-8') as f:
    json.dump(site_coords, f, ensure_ascii=False, indent=2)

print("Сгенерированные координаты записаны в:", OUT_DIR / "site_coords_suggested.json")
print("Таблица метрик сохранена в:", OUT_DIR / "site_metrics.csv")
print("\nЕсли хотите точнее — отредактируйте файл site_coords_suggested.json, подставив реальные пиксельные координаты, и затем выполните Ячейку B.")
```

```python
# ---------- Ячейка B: отрисовка карты с наложенными метриками (использует site_coords_suggested.json) ----------
# Вставьте эту ячейку сразу после Ячейки A и выполните. При необходимости сначала отредактируйте results_models/profile_analysis/map_overlay/site_coords_suggested.json

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
from PIL import Image
import json
from pathlib import Path
import numpy as np

MAP_PATH = "Карта.jpg"
OUT_DIR = Path("results_models/profile_analysis/map_overlay")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Загрузим метрики и координаты
site_metrics = pd.read_csv(OUT_DIR / "site_metrics.csv")
with open(OUT_DIR / "site_coords_suggested.json", 'r', encoding='utf-8') as f:
    site_coords = json.load(f)

# Проверим соответствие
sites_in_metrics = set(site_metrics['site'].tolist())
sites_in_coords = set(site_coords.keys())
missing = sites_in_metrics - sites_in_coords
if missing:
    print("ВНИМАНИЕ: для некоторых участков нет координат (используйте site_coords_suggested.json для ручного редактирования):", missing)
    # Для отсутствующих добавим автоматические координаты вдоль правой части карты (чтобы они нарисовались)
    img_tmp = Image.open(MAP_PATH).convert('RGBA')
    W, H = img_tmp.size
    x_auto = int(W * 0.55)
    ys = np.linspace(int(H*0.12), int(H*0.88), len(missing)).astype(int)
    for s, y in zip(missing, ys):
        site_coords[s] = (x_auto+20, int(y))

# Загрузим карту
img = Image.open(MAP_PATH).convert('RGBA')
W, H = img.size
fig, ax = plt.subplots(figsize=(8, 12))
ax.imshow(img)
ax.set_axis_off()

# Подготовим цвета: используем mean_delta (если положительное — намывание, отрицательное — размывание)
vals = site_metrics['mean_delta'].values
vmin, vmax = np.nanpercentile(vals, 5), np.nanpercentile(vals, 95)
# если vmin == vmax, немного расширим диапазон чтобы colormap работал
if np.isclose(vmin, vmax):
    vmin -= 1e-3
    vmax += 1e-3
norm = colors.TwoSlopeNorm(vmin=vmin, vcenter=0.0, vmax=vmax)  # центр 0: размывание <0, намывание >0
cmap = cm.get_cmap('RdBu_r')  # красный = намывание, синий = размывание (инвертируется _r при желании)

# отрисуем точки
for _, row in site_metrics.iterrows():
    site = row['site']
    mean_delta = row['mean_delta']
    n_prof = int(row['n_profiles'])
    # координаты
    x, y = site_coords.get(site, (int(W*0.55), int(H*0.5)))
    # marker size — базируется на числе профилей
    size = 50 * (1 + np.log1p(n_prof))
    color = cmap(norm(mean_delta))
    ax.scatter([x], [y], s=size, color=color, edgecolor='k', linewidth=0.4, zorder=20)
    # подпись: название участка и значение
    txt = f"{site}\n{mean_delta:.2f} (n={n_prof})"
    ax.text(x+8, y-6, txt, fontsize=8, va='top', zorder=21, bbox=dict(facecolor='white', alpha=0.6, pad=1, edgecolor='none'))

# Добавим легенду-карту цветов
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array(vals)
cbar = fig.colorbar(sm, ax=ax, fraction=0.03, pad=0.02)
cbar.set_label("Среднее изменение ИПН (первый прогнозный год) — положительное = намывание, отрицательное = размывание")

plt.title("Наложение прогнозов абразии по участкам на карту")
plt.tight_layout()

out_png = OUT_DIR / "map_abrasion_by_site_overlay.png"
fig.savefig(out_png, dpi=200, bbox_inches='tight')
plt.show()
print("Сохранено в:", out_png)
print("Если подписи/позиции нужно подправить — отредактируйте файл:", OUT_DIR / "site_coords_suggested.json", "затем снова выполните эту ячейку.")
```
